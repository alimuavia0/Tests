# Accumulators (Math)

### Q06 Calculate the Sum of all numbers 1 to 50

```python
sum = 0
for i in range(1,50):
    sum += i
print(sum)
```

### Q07 Calculate the product of all numbers from 1 to 5, (Factorial 5)

```python
fc = 1
print('(Factorial 5)')
for i in range(1,6):
    fc *= i
    
    print(fc)
```

### Q08 Sum of all list  elements

```python
list = [10,20,30,40]
sum = 0
for i in list:
    sum += i

print('Sum:',sum)
```

### Q09 Loop that counts how many numbers between 1 and 100 are divisible by 7.

```python
sum = []
for i in range(1,100):
    if i % 7==0 :
       sum.append(i)
       sm = len(sum)
print('Sum:',sm)
           
```

### Q15 Reverse a list[1,2,3,4] using a loop and store it in a new list

```python
list = [1,2,3,4]
Nlist = []
for i in list[::-1]:
    Nlist.append(i)
print(Nlist)
```
